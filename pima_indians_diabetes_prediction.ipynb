```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_data = pd.read_csv('diabetes.csv')
```

```python
csv_data.info()
```

```python
csv_data.isnull().count()
```

```python
data = csv_data[['Glucose', 'Insulin', 'BMI', 'Age', 'DiabetesPedigreeFunction', 'Outcome']]
```

```python
data.head()
```

```python
x = data[['Glucose', 'Insulin', 'BMI', 'Age', 'DiabetesPedigreeFunction']]
```

```python
y = data[['Outcome']]
```

```python
y = np.ravel(y)
```

```python
##############################################################################################
##############################################################################################
##############################################################################################
##############################################################################################
#                  Pima Indians Diabetes Database using Logistic Regression                  #
##############################################################################################
##############################################################################################
##############################################################################################
##############################################################################################
```

```python
# train test data splitting
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.8)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# importing model
from sklearn.linear_model import LogisticRegression
```

```python
# Model
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
y_predicted
```

```python
y_test
```

```python
accuracy = accuracy_score(y_test, y_predicted)*100
```

```python
accuracy
```
